```python
import json
tree_json_path = '../microenvironment/tree_yzx.json'
with open(tree_json_path, 'r') as file:
    tree_json = json.load(file) 
```

```python
brain_to_id = {}
id_to_brain = {}
brain_id_path = {}
for i in range(len(tree_json)):
    bid = tree_json[i]['id']
    acronym = tree_json[i]['acronym']
    p = tree_json[i]['structure_id_path']
    if bid not in brain_to_id:
        brain_to_id[bid] = acronym
        id_to_brain[acronym] = bid
    if bid not in brain_id_path:
        brain_id_path[bid] = p 
```

```python
range_r = [0,14,14,23,23,26,0,26]
for k in range(4):
    np.random.seed(42)
    M1 = np.corrcoef(mean_proj_r)[:,range_r[2*k]:range_r[2*k+1]][range_r[2*k]:range_r[2*k+1],:]
    M2 = np.corrcoef(match_glabel_p.T)[:,range_r[2*k]:range_r[2*k+1]][range_r[2*k]:range_r[2*k+1],:]

    N = M1.shape[0]
    triu_indices = np.triu_indices(N,k=1)
    v1 = M1[triu_indices]
    v2 = M2[triu_indices]

    obs_corr,_ = pearsonr(v1,v2)

    perm_corrs = np.zeros(1000)

    for i in range(1000):
        perm_indices = np.random.permutation(N)
        
        pm2 = M2.copy()
        pm2 = pm2[perm_indices,:]
        pm2 = pm2[:,perm_indices]
        
        pm_v2 = pm2[triu_indices]
        
        perm_corr,_ = pearsonr(v1,pm_v2)
        
        perm_corrs[i] = perm_corr 
        
    p_value = round(np.mean(np.abs(perm_corrs) >= np.abs(obs_corr)),6)
    print(p_value,obs_corr)
```

```python
### function match
function_df = pd.read_excel('../panoramic_view/brain_regions_functions.xlsx')
```

```python
uniq_regions = np.unique(full_class_infoTable['Soma_region'])
```

```python
unique_funcs = []
region_func = {}
func_region = {}
for i in range(len(uniq_regions)):
    r = uniq_regions[i]
    if r not in list(function_df['brain_region']):
        continue
    row_id = function_df[function_df['brain_region'] == r].index[0]
    fs = function_df['Function'].iloc[row_id]
    fs = fs.split(',')
    region_func[r] = fs
    for j in fs:
        if j not in unique_funcs:
            unique_funcs.append(j)
            func_region[j] = [r]
        else:
            func_region[j].append(r)
```

```python
uniq_r = np.unique(full_class_infoTable['Soma_region'])
```

```python
region_func_all = {}
func_region_all = {}
unique_funcs_all = []
for i in range(len(uniq_r)):
    r = uniq_r[i]
    if r in list(function_df['brain_region']):
        row_id = function_df[function_df['brain_region'] == r].index[0]
        fs = function_df['Function'].iloc[row_id]
        fs = fs.split(',')
        region_func_all[r] = fs
        for j in fs:
            if j not in unique_funcs_all:
                unique_funcs_all.append(j)
                func_region_all[j] = [r]
            else:
                func_region_all[j].append(r)
```

```python
func_MMB_loc = np.zeros((len(uniq_nlp),len(unique_funcs_all)))
func_MMB_loc2 = np.zeros((len(uniq_nlp),len(unique_funcs_all)))
for i in range(len(uniq_nlp)):
    names = nlp_table[(nlp_table['nlp'] == uniq_nlp[i])]['name'].tolist()
    rows = nlp_table[(nlp_table['nlp'] == uniq_nlp[i])]
    u_regions = np.unique(rows['region'])
    
    for j in u_regions:
        n = len(rows[rows['region']==j])
        if j not in region_func_all:
            continue
        fs = region_func_all[j]
        for f in fs:
            f_id = unique_funcs_all.index(f)
            func_MMB_loc[i,f_id] += n
            func_MMB_loc2[i,f_id] += n
```

```python
import math
```

```python
N = len(full_class_infoTable)
for i in range(len(uniq_nlp)):
    m = len(nlp_table[nlp_table['nlp'] == uniq_nlp[i]])
    for j in range(len(unique_funcs_all)):
        fr = func_region[unique_funcs_all[j]]
        nf = len(full_class_infoTable[full_class_infoTable['Soma_region'].isin(fr)])
        if func_MMB_loc[i,j] == 0:
            s= 0
        else:
            s = math.log2(func_MMB_loc[i,j]*N/(m*nf))
        func_MMB_loc2[i,j] = func_MMB_loc2[i,j]/m
        func_MMB_loc[i,j] = s
```

```python
yname = []
prefix = ['F','T','A','B']
for i in uniq_nlp:
    tmp_name = ''
    for j in range(len(i)):
        tmp_name = tmp_name + prefix[j]+str(int(i[j])+1)
    yname.append(tmp_name) 
```

```python
from scipy.cluster.hierarchy import fcluster,dendrogram,linkage
import scipy.cluster.hierarchy as sch 
```

```python
### target based 
func_MMB_target = np.zeros((len(uniq_nlp),len(unique_funcs_all)))
func_MMB_target2 = np.zeros((len(uniq_nlp),len(unique_funcs_all)))
for i in range(len(uniq_nlp)):
    names = nlp_table[(nlp_table['nlp'] == uniq_nlp[i])]['name'].tolist()
    rows = nlp_table[(nlp_table['nlp'] == uniq_nlp[i])]
    proj_rows = full_proj2.loc[names]
    for j in range(len(proj_rows)):
        for k in proj_rows.columns:
            if 'ipsi' in k:
                r = k.split('ipsi_')[1]
            else:
                r = k.split('contra_')[1]
            if r in region_func_all:
                proj_s = proj_rows[k].iloc[j]
                w = proj_s/np.sum(proj_rows.iloc[j])
                fs = region_func_all[r]
                for f in fs:
                    f_id = unique_funcs_all.index(f)
                    func_MMB_target[i,f_id] = func_MMB_target[i,f_id] + w
                    func_MMB_target2[i,f_id] = func_MMB_target2[i,f_id] + w
```

```python
N = len(full_class_infoTable)
for i in range(len(uniq_nlp)):
    nlp_l = len(nlp_table[(nlp_table['nlp'] == uniq_nlp[i])])
    for j in range(len(unique_funcs_all)):
        s_tmp = 0
        fr = func_region[unique_funcs_all[j]]
        for k in fr:
            for m in ['ipsi_'+k, 'contra_'+k]:
                if m not in (list(full_proj2.columns)):
                    continue
                rows = full_proj2[full_proj2[m]>0]
                for n in range(len(rows)):
                    s_tmp += full_proj2[m].iloc[n]/np.sum(full_proj2.iloc[n])
        if func_MMB_target[i,j] == 0:
            s= 0
        else:
            s = math.log2(func_MMB_target[i,j]*N/(s_tmp*nlp_l))
        func_MMB_target[i,j] = s
        func_MMB_target2[i,j] = func_MMB_target2[i,j]/nlp_l
```
